```python
#데이터 정제. 너무 벗어난 값은 없애버리는 작업
import numpy as np

def outliers_iqr(data):
    q1, q3=np.percentile(data,[25,75])
    iqr=q3-q1
    lower_bound = q1-(iqr*1.5)
    upper_bound = q3+(iqr*1.5)
    return np.where((data>upper_bound)|(data<lower_bound))

outliers= outliers_iqr(bike_data.Distance)
```

```python
np.mean(m_gu.Distance)
```

```python
#두 구의 평균 값이 다른데 이게 우연히 발생한 차이인지, 통계적으로 발생한 차이인지 비교필요

stats.ttest_ind(y_gu.Distance, m_gu.Distance, equal_var = True)
#마지막에 e-05. e 가 오일러상수라고함. 
#자연과학적 표기방법으로 0.05보다 꽤 많이 작은 값의 수치.
#즉, pvalue가 0.05보다 작다고 이해할 수 있음
```

# 5개 구 비교하기
## 분산분석의 단계
1. 등분산 분석
- 여러 그룹의 분산이 같은가 확인
- 가설검정 방법대로 귀무가설정. 모든 그룹의 분산은 같다
- p value가 유의수준보다 작으면 기각
- 귀무가설의 기각이란? 모든 그룹의 분산이 같은 것은 아니다
- 모든 그룹이 등분산임을 가정할 수 있으면 분산분석, 아니면 다른 방법으로 수행

2. 분산 분석
- 귀무가설: 모든 그룹의 평균이 같다
- 대립가설: 어떤 그룹의 평균은 같지 않다
- pvalue가 0.05보다 작으면 귀무가설 기각. 어떤 그룹의 평균은 같지 않다
- 그 어떤 그룹이 어떤 그룹인지 알 수 없음. 그래서 사후 분석하는 것.

3. 사후 분석
- 그룹 별 평균을 비교해 평균이 같은 그룹과 다른 그룹을 구분하여 알려줌

```python
#Bartlett's 등분산분석
s_gu = bike_data2[bike_data2.Gu == '서초구']
d_gu = bike_data2[bike_data2.Gu == '동대문구']
e_gu = bike_data2[bike_data2.Gu == '은평구']
```

```python
from scipy import stats

stats.bartlett(y_gu.Distance, m_gu.Distance, s_gu.Distance, d_gu.Distance, e_gu.Distance)
'''
pvalue가 e-86이라는 거는 0.05보다 매우 작다는 것. 즉 귀무가설 기각
등분산이 나오지 않았을 때는 
1. 비모수적 분석방법을 사용
    :통계량에 근거한 추론 방법을 사용하지 않는 것
2. 데이터에 보완점 검토
    : 데이터가 충분하지 않은 경우
'''
```

```python
#등분산이라고 가정하고 분산분석 진행
stats.f_oneway(y_gu.Distance, m_gu.Distance, s_gu.Distance, d_gu.Distance, e_gu.Distance)

#pvalue가 0.05보다 작아서 귀무가설 기각
```

```python
plot_data = [y_gu.Distance, m_gu.Distance, s_gu.Distance, d_gu.Distance, e_gu.Distance]
plt.boxplot(plot_data, showfliers=False)
plt.show()

#showfliers 는 outlier 표시 않고 전체 데이터 대상으로 그림
```

```python
#사후분석 Tukey HSD Honestly Significant Difference. 
#평균이 같은 그룹과 다른 그룹을 나누는 방법. statmodels 패키지 필요

from statsmodels.stats.multicomp import pairwise_tukeyhsd

hsd = pairwise_tukeyhsd(bike_data2.Distance, bike_data2.Gu)
hsd.summary()

#padj값이 0.05보다 작으면 True, 크면 False
#즉, 영등포 은평구만 이동거리 평균이 같고 다른 모든 구는 같지 않다
```
